```python
def print_dictionary(some_dictionary):
    for key in some_dictionary:
        print(key,":\t", some_dictionary[key])
```

```python
my_dictionary = { 'cs': 'computer science', 'chem':'chemistry','bio':'biology','hist':'history'}

print_dictionary(my_dictionary)
```

```python
import pandas as pd

# Read the csv file test_data2.csv
my_data = pd.read_csv("assets/test_data2.csv")

# Print information abou the returned object (which is a pandas DataFrame object)
print(my_data.info(),"\n\n")

# Or use the isinstance function
if isinstance(my_data, pd.core.frame.DataFrame):
    print("my_data is a DataFrame object")
else:
    print("my_data is of type", type(my_data))
```

```python
print(my_data.head(4)) # Print the top 4 rows
print("\n")
print(my_data.tail(2)) # Print the bottom 2 rows
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

x_value = my_data["EmployeeID"]
y_value = my_data["Salary"]

# Add axes Labels and plot title
plt.xlabel('EmployeeID')
plt.ylabel('Salary')
plt.title('My data plot')

# Rotate the x-axis labels by 90% so texts don't overlap on top of each other
plt.xticks(rotation=90)

# plot x values vs. y values
plt.plot(x_value,y_value)
plt.show()
```
