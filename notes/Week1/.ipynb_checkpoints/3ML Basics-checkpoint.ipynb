```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Generate random samples using numpy
# Let's generate 100 random numbers from a normal distribution
random_samples = np.random.randn(100)

# Display the random samples
print("Random Samples:")
print(random_samples)
# Create a simple DataFrame using pandas
df = pd.DataFrame({
    'Samples': random_samples
})

# Display the first few rows of the DataFrame
print("\nDataFrame displaying the random samples:")
print(df.head())
```

```python
# Plotting the random samples using matplotlib
plt.figure(figsize=(10, 6))
plt.plot(random_samples, label='Random Samples')
plt.title("Random Samples Plot")
plt.xlabel("Index")
plt.ylabel("Value")
plt.legend()
plt.show()
```

```python
# Demonstrating overfitting vs. underfitting with a simple polynomial regression example
# Generate synthetic data for this purpose
np.random.seed(42)
x = np.linspace(-3, 3, 100)
y = 0.5 * x ** 3 - x + 2 + np.random.randn(100) * 2  # True underlying function with noise

# Plot the original data
plt.figure(figsize=(10, 6))
plt.scatter(x, y, color='blue', label='Original Data')
```

```python
# Fit different models to demonstrate underfitting, overfitting, and balanced fitting
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import make_pipeline

# Underfitting - Linear model (degree 1)
model_underfit = make_pipeline(PolynomialFeatures(degree=1), LinearRegression())
model_underfit.fit(x.reshape(-1, 1), y)
y_pred_underfit = model_underfit.predict(x.reshape(-1, 1))
```

```python
# Overfitting - Very high-degree polynomial model (degree 15)
model_overfit = make_pipeline(PolynomialFeatures(degree=15), LinearRegression())
model_overfit.fit(x.reshape(-1, 1), y)
y_pred_overfit = model_overfit.predict(x.reshape(-1, 1))

# Balanced fit - Reasonable degree polynomial model (degree 3)
model_balanced = make_pipeline(PolynomialFeatures(degree=3), LinearRegression())
model_balanced.fit(x.reshape(-1, 1), y)
y_pred_balanced = model_balanced.predict(x.reshape(-1, 1))
```

```python
# Plot the results
plt.plot(x, y_pred_underfit, color='red', label='Underfitting (Linear Model)')
plt.plot(x, y_pred_overfit, color='green', label='Overfitting (Degree 15 Polynomial)')
plt.plot(x, y_pred_balanced, color='orange', label='Balanced Fit (Degree 3 Polynomial)')
plt.title("Overfitting vs. Underfitting")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```
