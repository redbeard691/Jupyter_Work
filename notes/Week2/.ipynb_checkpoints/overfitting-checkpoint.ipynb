```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
from sklearn.metrics import mean_squared_error

# 1. Generate Synthetic Data
np.random.seed(42)
X = np.linspace(-3, 3, 100).reshape(-1, 1)
y = X**3 - 5*X**2 + np.random.normal(0, 2, X.shape)

# 2. Fit Models
# Linear Model
linear_model = LinearRegression()
linear_model.fit(X, y)

# Polynomial Model of Degree 3
poly3 = PolynomialFeatures(degree=3)
poly3_model = make_pipeline(poly3, LinearRegression())
poly3_model.fit(X, y)

# Polynomial Model of Degree 15
poly15 = PolynomialFeatures(degree=15)
poly15_model = make_pipeline(poly15, LinearRegression())
poly15_model.fit(X, y)

# 3. Visualization
plt.figure(figsize=(16, 12))

# Plot Original Data
plt.subplot(2, 2, 1)
plt.scatter(X, y, color='black', s=10)
plt.title("Original Data")
plt.xlabel("Feature")
plt.ylabel("Target")

# Plot Linear Model Predictions
plt.subplot(2, 2, 2)
X_range = np.linspace(-3, 3, 100).reshape(-1, 1)
y_pred_linear = linear_model.predict(X_range)
plt.scatter(X, y, color='black', s=10, label="Data")
plt.plot(X_range, y_pred_linear, color='blue', label="Linear Model")
plt.title("Linear Model")
plt.xlabel("Feature")
plt.ylabel("Target")
plt.legend()

# Plot Polynomial Model (Degree 3) Predictions
plt.subplot(2, 2, 3)
y_pred_poly3 = poly3_model.predict(X_range)
plt.scatter(X, y, color='black', s=10, label="Data")
plt.plot(X_range, y_pred_poly3, color='green', label="Polynomial Model (Degree 3)")
plt.title("Polynomial Model (Degree 3)")
plt.xlabel("Feature")
plt.ylabel("Target")
plt.legend()

# Plot Polynomial Model (Degree 15) Predictions
plt.subplot(2, 2, 4)
y_pred_poly15 = poly15_model.predict(X_range)
plt.scatter(X, y, color='black', s=10, label="Data")
plt.plot(X_range, y_pred_poly15, color='red', label="Polynomial Model (Degree 15)")
plt.title("Polynomial Model (Degree 15)")
plt.xlabel("Feature")
plt.ylabel("Target")
plt.legend()

plt.tight_layout()
plt.show()
```
